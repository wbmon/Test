```python
# Свободное от уроков время ученики 10 класса любят проводить за логическими играми.
# Однажды два одноклассника Сергей и Влад решили придумать собственную игру,
# рассчитанную на двух участников.
# По правилам первый игрок загадывает произвольное трёхзначное натуральное число.
# Второй игрок из цифр этого числа образует максимальное и минимальное двузначные числа,
# а затем находит их разность. Если разность полученных двузначных чисел равна 10,
# то побеждает первый игрок, в противном случае выигрывает его соперник.
# Ребята задались вопросом: с какими числами на отрезке [500; 600] выиграет первый игрок?
# Найдите сумму этих чисел и запишите её в ответ.

vy = 0
for n in range (500,601):
    
    l = sorted([n//100, n%10, n%100//10])
    if l[0] != 0:
        res = l[0] * 10 + l[1]
    elif l[1] == 0:
        res = l[2] * 10 + l[1]
    else:
        res = l[1] * 10 + l[0]
    min = res
    max = l[2] * 10 + l[1]
    if abs(min - max) == 10: vy = vy + n
#     print(min, max, n, abs(min - max))
print(vy)
```

```python
# На уроке информатики ученики получили шифрованное сообщение.
# Учитель информатики подсказал ребятам, что ключом к шифровке
# является сумма всех уникальных чисел на отрезке [10; 1000000].
# Число называется уникальным, если сумма его цифр, возведённых
# в степень, равную количеству цифр числа, совпадает с самим числом.
# Например, 371 = 3^3 + 7^3 + 1^3. Помогите ребятам получить ключ к шифрованному сообщению.

def is_unique(num):
   n = len(str(num))
   sum_digits = sum(int(digit) ** n for digit in str(num))
   return num == sum_digits
unique_sum = 0
for num in range(10, 1000001):
   if is_unique(num):
       unique_sum += num
print(unique_sum)
```

```python
# Перевод числа в любую систему счисления

def convert_to(number, base, upper=False):
    digits = '0123456789abcdefghijklmnopqrstuvwxyz'
    if base > len(digits): return None
    result = ''
    while number > 0:
        result = digits[number % base] + result
        number //= base
    return result.upper() if upper else result
print(convert_to(987654321, 16))
```

### A. Доли простых чисел
Последовательность натуральных чисел от 1 до N включительно разбили на части. Каждая часть имеет длину M. Получилось N/M подпоследовательностей с числами в диапазоне [1, M], [M+1, 2M] и т. д. В каждой части найдите долю простых чисел.
Формат ввода В первой строке входных данных записано натуральное число N (1 ≤ N ≤ 100 000). Вторая строка содержит натуральное число M (1 ≤ M ≤ N). Гарантируется, что N делится без остатка на M.
Формат вывода Выведите N/M строк. В k-й строке выведите долю простых чисел в последовательности [(k-1)*M+1, k*M] с точностью до 4 знаков после запятой. Решение

```python
# Решето Эратосфена - это алгоритм, позволяющий найти все простые числа в диапазоне [1; n].
# Отмечаем нулями все ячейки списка arr размера n+1, у которых индексы не простые числа, используя алгоритм Решето Эратосфена.
def calc_erato():
    arr[0] = 0
    arr[1] = 0
    i = 2
    while i * i < len(arr):
        if arr[i]:
            for j in range(i * i, len(arr), i):
                arr[j] = 0
        i += 1
n = int(input())
m = int(input())
count = 0

# Изначально список заполнен единицами
arr = [1 for i in range(n + 1)]
calc_erato() # Отмечаем все простые числа в списке arr
for i in range(1, n + 1): # Пробегаемся по всему списку и посчитаем количество простых для каждого промежутка
    if arr[i]:
        count += 1
    if i % m == 0:
        print(count / m)
        count = 0
```
